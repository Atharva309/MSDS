Title: Exploring NYPD Shootings Incidents: A Data Analysis using R Notebook

Slide 1: Introduction
- **Objective:** Analyzing NYPD Shootings Incidents from 2006 to 2021 using R Notebook.
- **Tools:** R Notebook, knitr, tidyverse, lubridate, ggplot2.
- **Key Code Settings:** `knitr::opts_chunk$set(echo = TRUE)` for code visibility.

Slide 2: Data Retrieval and Cleaning
- **Data Source:** NYPD Shootings Incidents dataset from NYC Open Data.
- **Libraries Used:** tidyverse, lubridate.
- **Data Cleaning Steps:**
  - Selected relevant columns.
  - Converted date and time columns to appropriate formats.
  - Created a new variable for shootings.
  - Extracted the year from the date.

Slide 3: Shootings by Borough (Figure - 1)
- **Visualization:** Bar chart showing total shootings by borough.
- **Insights:**
  - Brooklyn has the highest number of shootings.
  - Staten Island has the lowest number.

Slide 4: Shootings by Year (Figure - 2)
- **Visualization:** Bar chart displaying total shootings per year.
- **Insights:**
  - Shootings peaked around 2011.
  - A gradual decline in recent years.

Slide 5: Shootings Trends Over Years (Figure - 3)
- **Visualization:** Line chart representing shootings trend over 2006-2021.
- **Insights:**
  - Overall decreasing trend in shootings.
  - Anomaly in 2021 (possible data incompleteness).

Slide 6: Borough-wise Shootings Over Time (Figure - 4)
- **Visualization:** Line chart depicting borough-wise shootings by year.
- **Insights:**
  - Brooklyn consistently has the highest shootings.
  - Other boroughs show varying trends.

Slide 7: Statistical Analysis
- **Analysis 1:** Proportion of total shootings in Brooklyn and Queens combined.
  - Result: 65% of total shootings occurred in Brooklyn and Queens.

- **Analysis 2:** Proportion of total shootings in Staten Island.
  - Result: Only 2.7% of total shootings occurred in Staten Island.

Slide 10: NYPD Shootings Per Day (Figure - 5)
- **Visualization:** Line chart illustrating the daily count of shootings from 2006 to 2021.
- **Insights:**
  - Peaks and valleys in shooting incidents over the years.
  - Highest shooting days around July 4th, 2023.

Slide 11: Shootings on July 5, 2023 (Figure - 6)
- **Visualization:** Line chart showing shootings on July 5, 2023, with a focus on the top two shooting days.
- **Insights:**
  - July 5, 2023, had the highest number of shootings.
  - The second-highest shootings occurred on July 4, 2023.

Slide 12: NYPD Shootings by the Time of Day (Figure - 7)
- **Visualization:** Line chart presenting the distribution of shootings throughout the 24-hour time system.
- **Insights:**
  - Shootings tend to be higher during the early morning hours.

Slide 13: Time of Day Analysis with Polynomial Regression (Figure - 8)
- **Model:** Polynomial regression model applied to analyze shootings per hour.
- **Model Insights:**
  - The model suggests a quadratic relationship between the hour of the day and shootings.
  - Residuals indicate a good fit to the data.
  - Shooting incidents are concentrated during specific hours.

Slide 14: Conclusion
- **Summary:** The analysis provides a comprehensive understanding of NYPD shootings incidents, considering temporal and spatial dimensions.
- **Key Findings:**
  - Peaks in shootings around specific days (e.g., July 4, 2023).
  - Concentration of shootings during certain hours of the day.
  - A decreasing trend in overall shootings from 2006 to 2021.

Slide 15: Next Steps
- **Recommendations:**
  - Investigate anomalies in 2021 and explore potential contributing factors.
  - Conduct further analysis to understand the temporal patterns in more detail.
  - Consider additional data sources to enrich the analysis.
