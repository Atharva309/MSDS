```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming you already have the DataFrame 'df' loaded with your data

# Group the data by 'datetime' and count the number of incidents for each date
incident_counts = df.groupby('datetime').size().reset_index(name='incident_count')

# Use .to_numpy() to convert datetime and incident_count to NumPy arrays
datetime_array = incident_counts['datetime'].to_numpy()
incident_count_array = incident_counts['incident_count'].to_numpy()

# Create the time series plot using matplotlib
plt.figure(figsize=(12, 6))
plt.plot(datetime_array, incident_count_array, marker='o', linestyle='-')
plt.title('Incidents Over Time')
plt.xlabel('Date')
plt.ylabel('Incident Count')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```
