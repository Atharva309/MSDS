#  Massive software hack exposes most Louisianans' OMV data, officials say
The leak is affecting most Louisianans with a state driver's license, state officials said Thursday. 

https://www.nola.com/news/politics/legislature/massive-software-hack-exposes-most-louisianans-omv-data/article_d16a549c-0bbe-11ee-9c01-7f6c2375dace.html

If you have a Louisiana driver's license, your personal information was likely exposed in a massive data breach caused by the hacking of third-party software used by the state's Office of Motor Vehicles, officials said Thursday.

OMV officials believe all Louisianans with a state-issued driver’s license, ID, or car registration have had some of their data made vulnerable by the cyber attacks, according to the Governor's Office of Homeland Security and Emergency Preparedness. That could include names, addresses, social security numbers, birthdates, height, eye color, driver's license numbers, vehicle registration and handicap placard information.

GOHSEP officials called the attack "unprecedented" in a news release. State officials didn't provide any more information but said they plan to hold a press conference Friday.

It's unclear exactly how many agencies in Louisiana were affected, they said. But the file-sharing software targeted at OMV, MOVEit, was at the root of a sweeping international breach that was hitting universities, companies and a number of U.S. government agencies Thursday, according to reports. 

The US Cybersecurity and Infrastructure Security Agency “is providing support to several federal agencies that have experienced intrusions affecting their MOVEit applications,” Eric Goldstein, the agency’s executive assistant director for cybersecurity, said in a statement to news outlets.

The list of non-governmental victims apparently affected was also ballooning Thursday and included Johns Hopkins University in Baltimore, The University of Georgia System, British Airways and the BBC, according to reports. Bloomberg News reported that Shell, the oil and gas giant with a presence in Louisiana, was a potential casualty.

MOVEit is a file transfer service used by government agencies and companies around the world. Some U.S. federal agencies use it to store sensitive data.

There are currently no signs that cyber attackers who breached the software in Louisiana have sold, used or released OMV data, GOHSEP officials said. The cyber attackers have not contacted state government, they said.

But all Louisianans should take steps to safeguard their identity, they said, like checking credit reports, changing passwords and other actions.

Federal authorities have not publicly identified any suspects in Thursday's hack.

But in recent days, federal cybersecurity officials discovered security vulnerabilities in MOVEit software that have allowed a group of Russian-speaking cybercriminals — called the CL0P ransomware gang — to steal data and demand ransom payments, a new government report says.

Louisiana OMV Commissioner Karen St. Germain did not immediately return a phone message Thursday.

Spokespeople for the Louisiana Department of Health and Department of Veterans Affairs did not immediately return emailed messages asking whether their agencies were affected by the hack.
