Titanic dataset, it contains information about passengers on the Titanic, including details such as their survival status, class, name, sex, age, number of siblings/spouses aboard (SibSp), number of parents/children aboard (Parch), ticket information, fare, cabin, and embarked location.

## Key attributes/dimensions of the data:
- PassengerId: Unique identifier for each passenger.
- Survived: Binary variable indicating whether the passenger survived (1) or not (0).
- Pclass: Ticket class (1st, 2nd, or 3rd).
- Name: Passenger's name.
- Sex: Gender of the passenger.
- Age: Age of the passenger.
- SibSp: Number of siblings/spouses aboard.
- Parch: Number of parents/children aboard.
- Ticket: Ticket number.
- Fare: Fare paid by the passenger.
- Cabin: Cabin number.
- Embarked: Port of embarkation (C = Cherbourg, Q = Queenstown, S = Southampton).

## Goals:
- Predictive Modeling: Develop models to predict survival based on the available features.
- Exploratory Data Analysis (EDA)

## Questions
- What factors contributed to the survival of passengers?
- How does the distribution of ages vary among different passenger classes?
- Are there any patterns or trends based on gender, class, or embarkation port?
- Can we identify any outliers or anomalies in the fare distribution?

## Existing relevant visualizations:

- Survival Rate by Class and Gender: Bar charts or stacked bar charts can be used to visualize survival rates based on passenger class and gender.
- Age Distribution: Histograms or kernel density plots to show the distribution of passenger ages.

## Critique of existing visualizations:
- Survival Rate by Class and Gender: This visualization is effective in comparing survival rates, but it could be enhanced by providing absolute numbers along with percentages.
- Age Distribution: Histograms are useful, but adding age group labels or overlaying the distribution with class information can provide more insights.
