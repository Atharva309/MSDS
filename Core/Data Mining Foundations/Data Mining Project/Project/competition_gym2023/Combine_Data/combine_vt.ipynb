```python
def rename_columns(dataframe):
    # Define the column name mapping
    column_name_mapping = {
        'Total': 'Average',
        'Score': 'Total',
        'd2': 'D2',
        'e2': 'E2',
        'nd2': 'ND2',
        'score2': 'Total2',
    }

    return dataframe.rename(columns=column_name_mapping, inplace=True)
```
