```python
import pandas as pd
import numpy as np
```

```python
iaa21 = pd.read_csv('../../Cleaning/qual_events/men/iaa2021.csv')
iaa22 = pd.read_csv('../../Cleaning/qual_events/men/iaa2022.csv')
iaa23 = pd.read_csv('../../Cleaning/qual_events/men/iaa2023.csv')
```

```python
combined_df = pd.concat([iaa21, iaa22, iaa23], ignore_index=True)
iaa24=combined_df.copy()
```

```python
import re

# Define a function to remove parentheses and their contents
def remove_parentheses(text):
    return re.sub(r'\([^)]*\)', '', text)

# Define the columns to clean
columns_to_clean = ['FX', 'PH', 'SR', 'VT', 'PB', 'HB', 'AA']  # Add other column names if needed

# Ensure that the selected columns are of string data type
iaa24[columns_to_clean] = iaa24[columns_to_clean].astype(str)

# Apply the function to the DataFrame
df_cleaned = iaa24.copy()
df_cleaned[columns_to_clean] = df_cleaned[columns_to_clean].applymap(remove_parentheses)

# Display the cleaned DataFrame
print(df_cleaned)
```

```python
iaa24=df_cleaned.copy()
```

```python
columns_to_convert_to_float = ['FX', 'PH', 'SR', 'VT', 'PB', 'HB', 'AA']
columns_to_convert_to_int = ['Rank']

# Convert columns to float
iaa24[columns_to_convert_to_float] = iaa24[columns_to_convert_to_float].apply(pd.to_numeric, errors='coerce')

# Convert 'Rank' column to int
iaa24['Rank'] = iaa24['Rank'].apply(pd.to_numeric, downcast='integer', errors='coerce')
```
