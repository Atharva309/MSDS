```python
# Define the qualified teams
qualified_teams = {
    'm': ['CHI', 'JPN', 'ENG'],
    'w': ['USA', 'ENG', 'CAN']
}

# Filter the DataFrame to get the qualified teams
qualified_teams_df = df[df['Country'].isin(qualified_teams['m'] + qualified_teams['w'])]

# Separate male and female qualified teams
male_teams_df = qualified_teams_df[(qualified_teams_df['Gender'] == 'm') & (qualified_teams_df['Country'].isin(qualified_teams['m']))]
female_teams_df = qualified_teams_df[(qualified_teams_df['Gender'] == 'w') & (qualified_teams_df['Country'].isin(qualified_teams['w']))]

# Print the qualified teams for each gender
print("Qualified Male Teams:")
print(male_teams_df['Country'].unique())

print("Qualified Female Teams:")
print(female_teams_df['Country'].unique())
```
