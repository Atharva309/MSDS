```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


# Display basic information about the DataFrame
print(df.info())

```

```python

# Summary statistics of numerical columns
print(df.describe())

# Check for missing values
print(df.isnull().sum())

```

```python

# Boxplot for 'D Score', 'E Score', and 'Total'
plt.figure(figsize=(12, 6))
sns.boxplot(data=df[['D Score', 'E Score', 'Total']])
plt.title('Boxplot of D Score, E Score, and Total')
plt.xlabel('Metrics')
plt.ylabel('Score')
plt.show()

# Countplot for 'Country' column
plt.figure(figsize=(12, 6))
sns.countplot(data=df, x='Country')
plt.title('Count of Gymnasts by Country')
plt.xlabel('Country')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()

# Pairplot for numerical columns
sns.pairplot(df[['D Score', 'E Score', 'Total', 'Pen.']])
plt.suptitle('Pairplot of Numerical Columns')
plt.show()

# Correlation heatmap
correlation_matrix = df[['D Score', 'E Score', 'Total', 'Pen.']].corr()
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```
