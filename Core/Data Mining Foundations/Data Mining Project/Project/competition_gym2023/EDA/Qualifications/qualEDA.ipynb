```python
import pandas as pd
import numpy as np
```

```python
df08 = pd.read_csv('../../Cleaning/qual2008/clean2008qual_team.csv')
```

```python
df08
```

```python
# remove 0's from total_rank column then check correlation betwen total rank and total score
```

```python
# Remove rows where "Rank" is 0 or less
df08r = df08[df08['total_rank'] > 0]
```

```python
df08r=df08r[df08r["total_score"]<100]
```

```python
import matplotlib.pyplot as plt

# Assuming you have a DataFrame with columns "total_rank" and "total_score"
# Replace df with your actual DataFrame if needed

# Create a scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(df08r["total_rank"], df08r["total_score"], alpha=0.5)
plt.title("Correlation between Total Rank and Total Score")
plt.xlabel("Total Rank")
plt.ylabel("Total Score")
plt.grid(True)

# Show the plot
plt.show()
```

```python
correlation = df08r["total_score"].corr(df08r["total_rank"])
correlation
```
