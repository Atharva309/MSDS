```python

# Extract columns ending with "exec" and "diff"
exec_columns = df.filter(like='_EXEC')
diff_columns = df.filter(like='_Diff')

# Filter values greater than 0 in the 'exec' columns
exec_columns = exec_columns[exec_columns > 0].dropna()

# Filter values greater than 0 in the 'diff' columns
diff_columns = diff_columns[diff_columns > 0].dropna()

# Summary statistics for the 'exec' columns
exec_summary_stats = exec_columns.describe()
print(exec_summary_stats)

print('\n')

# Summary statistics for the 'diff' columns
diff_summary_stats = diff_columns.describe()
print(diff_summary_stats)

# Visualize the distribution of 'exec' columns using histograms
plt.figure(figsize=(12, 8))
exec_columns.hist(bins=20, edgecolor='k', alpha=0.7)
plt.title('Histograms of Columns Ending with "_EXEC" (Values > 0)')
plt.tight_layout()

# Visualize the distribution of 'diff' columns using histograms
plt.figure(figsize=(12, 8))
diff_columns.hist(bins=20, edgecolor='k', alpha=0.7)
plt.title('Histograms of Columns Ending with "_Diff" (Values > 0)')
plt.tight_layout()

# Pairwise correlation matrix for 'exec' columns
exec_correlation_matrix = exec_columns.corr()

# Pairwise correlation matrix for 'diff' columns
diff_correlation_matrix = diff_columns.corr()

# Visualize correlation using heatmaps for 'exec' columns
plt.figure(figsize=(12, 8))
sns.heatmap(exec_correlation_matrix, cmap='coolwarm', annot=True)
plt.title('Correlation Heatmap for Columns Ending with "_EXEC" (Values > 0)')

# Visualize correlation using heatmaps for 'diff' columns
plt.figure(figsize=(12, 8))
sns.heatmap(diff_correlation_matrix, cmap='coolwarm', annot=True)
plt.title('Correlation Heatmap for Columns Ending with "_Diff" (Values > 0)')

# Show plots
plt.show()
```
