```python
import pandas as pd
import numpy as np
```

```python
df20 = pd.read_csv('../../Cleaning/olympics/womens/clean2020Olympics_women\'s_artistic_team_all-around.csv')
df16 = pd.read_csv('../../Cleaning/olympics/womens/clean2016Olympics_women\'s_artistic_team_all-around.csv')
df12 = pd.read_csv('../../Cleaning/olympics/womens/clean2012Olympics_women\'s_artistic_team_all-around.csv')
```

```python
combined_df = pd.concat([df12, df16, df20], axis=0, ignore_index=True)
```

```python
combined_df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
medals = combined_df[combined_df['Rank'].isin([1.0, 2.0, 3.0])]
```

```python
df = medals.copy()
```

```python
# Plotting using Seaborn
sns.set(style="whitegrid")

# Bar plot of 'Total' by 'Country' for each year
plt.figure(figsize=(12, 6))
sns.barplot(data=df, x='Total', y='Country', hue='year')
plt.title('Total Scores by Country')
plt.xlabel('Total Score')
plt.ylabel('Country')
plt.show()
```

```python
# Scatter plot of 'vt' vs 'ub'
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='vt', y='ub', hue='year')
plt.title('vt vs ub')
plt.xlabel('vt')
plt.ylabel('ub')
plt.show()
```

```python


# Box plot of 'bb' by 'year'
plt.figure(figsize=(8, 6))
sns.boxplot(data=df, x='year', y='bb')
plt.title('Box Plot of bb by Year')
plt.xlabel('Year')
plt.ylabel('bb')
plt.show()
```

```python
# Box plot of 'bb' by 'year'
plt.figure(figsize=(8, 6))
sns.boxplot(data=df, x='year', y='vt')
plt.title('Box Plot of vt by Year')
plt.xlabel('Year')
plt.ylabel('vt')
plt.show()
```

```python
# Box plot of 'bb' by 'year'
plt.figure(figsize=(8, 6))
sns.boxplot(data=df, x='year', y='ub')
plt.title('Box Plot of ub by Year')
plt.xlabel('Year')
plt.ylabel('ub')
plt.show()
```

```python
# Box plot of 'bb' by 'year'
plt.figure(figsize=(8, 6))
sns.boxplot(data=df, x='year', y='fx')
plt.title('Box Plot of fx by Year')
plt.xlabel('Year')
plt.ylabel('fx')
plt.show()
```
