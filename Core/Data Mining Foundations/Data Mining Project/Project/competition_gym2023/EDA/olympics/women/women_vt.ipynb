```python
import pandas as pd
import numpy as np
```

```python
dff = pd.read_csv('../../../Cleaning/olympics/womens/cleaned_womens2012to2020vt')
```

```python
dff
```

```python
dff.info()
```

```python
# Define a mapping dictionary from full country names to abbreviations
country_mapping = {
    'Romania': 'ROU',
    'United States': 'USA',
    'Russia': 'RUS',
    'Germany': 'GER',
    'Dominican Republic': 'DOM',
    'Canada': 'CAN'
}

# Use the replace function to replace the full country names with abbreviations
dff['Nation'] = dff['Nation'].replace(country_mapping)
```

```python
dff["Nation"].unique()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
medals = dff[dff['Rank'].isin([1.0, 2.0, 3.0])]
```

```python
medals
```

```python
df = medals.copy()
```

```python
df["Nation"].value_counts()
```

```python
# Box plot of 'bb' by 'year'
plt.figure(figsize=(8, 6))
sns.boxplot(data=df, x='year', y='vt1_D Score')
plt.title('Box Plot of vt by Year')
plt.xlabel('Year')
plt.ylabel('vt')
plt.show()
```

```python
# Box plot of 'bb' by 'year'
plt.figure(figsize=(8, 6))
sns.boxplot(data=df, x='year', y='vt1_E Score')
plt.title('Box Plot of vt by Year')
plt.xlabel('Year')
plt.ylabel('vt')
plt.show()
```

```python
df["avg_d"] = (df["vt2_D Score"] + df["vt1_D Score"])/2
```

```python
df["avg_e"] = (df["vt2_E Score"] + df["vt1_E Score"])/2
```

```python
# Box plot of 'bb' by 'year'
plt.figure(figsize=(8, 6))
sns.boxplot(data=df, x='year', y='avg_d')
plt.title('Box Plot of vt by Year')
plt.xlabel('Year')
plt.ylabel('vt')
plt.show()
```

```python
# Box plot of 'bb' by 'year'
plt.figure(figsize=(8, 6))
sns.boxplot(data=df, x='year', y='avg_e')
plt.title('Box Plot of vt by Year')
plt.xlabel('Year')
plt.ylabel('vt')
plt.show()
```

```python
# Box plot of 'bb' by 'year'
plt.figure(figsize=(8, 6))
sns.boxplot(data=df, x='year', y='Total')
plt.title('Box Plot of vault by Year')
plt.xlabel('Year')
plt.ylabel('Vault')
plt.show()
```
