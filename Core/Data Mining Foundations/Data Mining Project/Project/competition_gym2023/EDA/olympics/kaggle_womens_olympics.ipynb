# Change in Age and Height over Time for Medalist and all Gymnasts

```python
import matplotlib.pyplot as plt

# Group the data by 'Year' and calculate the mean age for each year
age_by_year = df.groupby('Year')['Age'].mean()

# Create a line plot
plt.figure(figsize=(10, 6))
age_by_year.plot(kind='line', marker='o')
plt.title('Average Age Over Time')
plt.xlabel('Year')
plt.ylabel('Average Age')
plt.grid(True)
plt.show()

```
