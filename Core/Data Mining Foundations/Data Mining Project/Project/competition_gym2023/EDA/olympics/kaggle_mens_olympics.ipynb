```python
import pandas as pd
import numpy as np
```

```python
df= pd.read_csv('../../kaggle-data/mens_olympics')
modern = pd.read_csv('../../kaggle-data/modern_mens_olympics')
```

```python
modern['Medal'].fillna('None', inplace=True)
```

```python
df['Medal'].fillna('None', inplace=True)
```

```python
medals = df[df["Medal"] != "None"]
```

```python
modern_medals = modern[modern["Medal"] != "None"]
```

```python
import matplotlib.pyplot as plt

# Group the data by 'Year' and calculate the mean age for each year
age_by_year = df.groupby('Year')['Age'].mean()

# Create a line plot
plt.figure(figsize=(10, 6))
age_by_year.plot(kind='line', marker='o')
plt.title('Average Age Over Time')
plt.xlabel('Year')
plt.ylabel('Average Age')
plt.grid(True)
plt.show()

```
