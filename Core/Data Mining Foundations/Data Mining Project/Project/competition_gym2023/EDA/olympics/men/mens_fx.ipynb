```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../../../Cleaning/olympics/mens/cleaned_mens2012to2020fx")
```

```python
df
```

```python
df['Total'] = df['Total'].str.replace('*', '').astype(float)
```

```python
df
```

```python
medals = df[df['Rank'].isin([1.0, 2.0, 3.0])]
```

```python
df = medals.copy()
```

```python
df["Nation"].value_counts()
```

```python
# Box plot of 'bb' by 'year'
plt.figure(figsize=(8, 6))
sns.boxplot(data=df, x='year', y='Total')
plt.title('Box Plot of Total by Year')
plt.xlabel('Year')
plt.ylabel('Total')
plt.show()
```

```python
df
```

```python
# Convert columns to NumPy arrays for plotting
year = df['year'].to_numpy()
d_avg = df['D Score'].to_numpy()
e_avg = df['E Score'].to_numpy()

# Create a line plot to show the change over time
plt.figure(figsize=(10, 6))
plt.plot(year, d_avg, label='d_avg', marker='o')
plt.plot(year, e_avg, label='e_avg', marker='o')
plt.xlabel('Year')
plt.ylabel('Average Score')
plt.title('Change Over Time of d_avg and e_avg')
plt.legend()
plt.grid(True)
plt.show()
```
