```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../../../Cleaning/olympics/mens/cleaned_mens2012to2020vt")
```

```python
df
```

```python
df.loc[16, 'vt2_E Score'] = df.loc[16, 'vt2_E Score'].replace('+C333', '')
```

```python
df.loc[16, 'Rank'] = 1
df.loc[17, 'Rank'] = 2
df.loc[18, 'Rank'] = 3
```

```python
# df["Total"] =(df["vt1_Vault Score"] + df["vt2_Vault Score"])/2
```

```python
df['vt2_E Score'] = pd.to_numeric(df['vt2_E Score'], errors='coerce', downcast='float')
```

```python
df.info()
```

```python
medals = df[df['Rank'].isin([1.0, 2.0, 3.0])]
```

```python
df = medals.copy()
```

```python
df["Nation"].value_counts()
```

```python
# Box plot of 'bb' by 'year'
plt.figure(figsize=(8, 6))
sns.boxplot(data=df, x='year', y='Total')
plt.title('Box Plot of Total by Year')
plt.xlabel('Year')
plt.ylabel('Total')
plt.show()
```

```python
# Calculate the correlation matrix
correlation_matrix = df[['vt1_D Score', 'vt1_E Score', 'vt1_Vault Score']].corr()

# Create a heatmap to visualize the correlations
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=.5)
plt.title('Correlation Heatmap - vt1_D Score, vt1_E Score, vt1_Vault Score')
plt.show()
```

```python
# Calculate the correlation matrix
correlation_matrix = df[['vt2_D Score', 'vt2_E Score', 'vt2_Vault Score']].corr()

# Create a heatmap to visualize the correlations
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=.5)
plt.title('Correlation Heatmap - vt2_D Score, vt2_E Score, vt2_Vault Score')
plt.show()
```

```python
df["d_avg"]= (df["vt1_D Score"] + df["vt2_D Score"])/2
```

```python
df
```

```python
df["e_avg"]= (df["vt1_E Score"] + df["vt2_E Score"])/2
```

```python
df
```

```python
# Convert columns to NumPy arrays for plotting
year = df['year'].to_numpy()
d_avg = df['d_avg'].to_numpy()
e_avg = df['e_avg'].to_numpy()

# Create a line plot to show the change over time
plt.figure(figsize=(10, 6))
plt.plot(year, d_avg, label='d_avg', marker='o')
plt.plot(year, e_avg, label='e_avg', marker='o')
plt.xlabel('Year')
plt.ylabel('Average Score')
plt.title('Change Over Time of d_avg and e_avg')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Convert columns to NumPy arrays for plotting
year = df['year'].to_numpy()
d_avg = df['d_avg'].to_numpy()
e_avg = df['e_avg'].to_numpy()

# Create a line plot for d_avg
plt.figure(figsize=(10, 6))
plt.plot(year, d_avg, label='d_avg', marker='o', color='blue')
plt.xlabel('Year')
plt.ylabel('Average d_avg Score')
plt.title('Change Over Time of d_avg')
plt.legend()
plt.grid(True)
plt.show()

# Create a line plot for e_avg
plt.figure(figsize=(10, 6))
plt.plot(year, e_avg, label='e_avg', marker='o', color='green')
plt.xlabel('Year')
plt.ylabel('Average e_avg Score')
plt.title('Change Over Time of e_avg')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Convert columns to NumPy arrays for plotting
year = df['year'].to_numpy()
total = df['Total'].to_numpy()

# Create a line plot for 'Total'
plt.figure(figsize=(10, 6))
plt.plot(year, total, label='Total', marker='o', color='purple')
plt.xlabel('Year')
plt.ylabel('Total Score')
plt.title('Change Over Time of Total Score')
plt.legend()
plt.grid(True)
plt.show()
```
