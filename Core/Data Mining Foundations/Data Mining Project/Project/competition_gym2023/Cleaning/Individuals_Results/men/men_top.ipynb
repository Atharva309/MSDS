```python
import pulp
import pandas as pd

# Define the athlete data
grouped_means = {
    'Asher Hong': {'FX': 13.116, 'PH': 12.728, 'SR': 14.3495, 'VT': 15.7175, 'PB': 14.5275, 'HB': 12.05, 'AA': 82.4885},
    'Colin Van': {'FX': 13.45, 'PH': 11.7, 'SR': 13.3, 'VT': 14.847, 'PB': 13.7, 'HB': 13.15, 'AA': 80.147},
    'Colt Walker': {'FX': 12.45, 'PH': 12.75, 'SR': 13.25, 'VT': 14.897, 'PB': 15.055, 'HB': 13.1, 'AA': 81.502},
    'Fred Richard': {'FX': 14.483, 'PH': 13.357, 'SR': 13.35, 'VT': 14.0, 'PB': 14.45, 'HB': 14.962, 'AA': 84.602},
    'Ian Lasic-Ellis': {'FX': 13.625, 'PH': 13.45, 'SR': 13.85, 'VT': 13.625, 'PB': 13.9, 'HB': 13.0275, 'AA': 81.4775},
    'Khoi Young': {'FX': 13.55, 'PH': 15.342, 'SR': 12.55, 'VT': 15.244, 'PB': 13.75, 'HB': 13.1, 'AA': 83.536},
    'Riley Loos': {'FX': 12.8, 'PH': 12.95, 'SR': 13.7765, 'VT': 14.4995, 'PB': 13.725, 'HB': 12.875, 'AA': 80.626},
    'Shane Wiskus': {'FX': 13.7, 'PH': 13.25, 'SR': 14.25, 'VT': 13.75, 'PB': 14.05, 'HB': 12.2, 'AA': 81.2},
    'Taylor Burkhart': {'FX': 13.2, 'PH': 13.6, 'SR': 12.95, 'VT': 14.594, 'PB': 14.15, 'HB': 12.7, 'AA': 81.194},
    'Vitaliy Guimaraes': {'FX': 14.25, 'PH': 14.35, 'SR': 13.6, 'VT': 14.45, 'PB': 13.9, 'HB': 13.4, 'AA': 83.95},
    'Yul Moldauer': {'FX': 13.9, 'PH': 13.873, 'SR': 14.115, 'VT': 13.825, 'PB': 15.332, 'HB': 12.45, 'AA': 83.495}
}

# Create a pandas DataFrame from the data
df = pd.DataFrame(grouped_means)

# Create a MILP problem
prob = pulp.LpProblem("Team_Selection", pulp.LpMaximize)

# Define decision variables
athletes = list(df.index)
apparatus = list(df.columns)
x = pulp.LpVariable.dicts("x", ((athlete, app) for athlete in athletes for app in apparatus), cat="Binary")

# Define the objective function
prob += pulp.lpSum(df.loc[athlete, app] * x[(athlete, app)] for athlete in athletes for app in apparatus)

# Add constraints
# Additional constraint to ensure unique athletes
for app in apparatus:
    prob += pulp.lpSum(x[(athlete, app)] for athlete in athletes) <= 1

# Add a constraint to select exactly 5 athletes
prob += pulp.lpSum(x[(athlete, app)] for athlete in athletes for app in apparatus) == 5

# Solve the problem with the CBC solver
prob.solve(pulp.PULP_CBC_CMD(msg=False))

# Print the selected team
selected_team = [(athlete, app) for athlete in athletes for app in apparatus if x[(athlete, app)].value() == 1]

print("Selected Team:")
for athlete, app in selected_team:
    print(f"{athlete} on {app}")
```
