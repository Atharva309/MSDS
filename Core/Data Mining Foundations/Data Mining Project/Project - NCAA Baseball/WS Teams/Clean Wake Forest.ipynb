```python
wakeb = wakeb.drop(columns=['IBB', 'Notes', 'Rk', 'SH', 'CS', 'SB', 'GDP'])
wakeb
```

```python
wakeb = wakeb.sort_values(by=['PA'], ascending=False)
wakeb
```

```python
wakeb
```

```python
wakeb = wakeb[:10]
wakeb = wakeb.drop(columns=['Name', 'Age'])
wakeb
```

```python
# Iterate through each row in the DataFrame
for index, row in wakeb.iterrows():
    # Divide each entry by the corresponding 'G' column entry
    for column in wakeb.columns:
        wakeb.at[index, column] = row[column] / row['G']
        
wakeb.head()
```

```python
wakeb.columns
```

```python
df2 = wakeb[['G', 'PA', 'AB', 'R', 'H', '2B', '3B', 'HR', 'RBI', 'BB', 'SO', 'BA',
       'OBP', 'SLG', 'OPS', 'TB', 'HBP', 'SF']].mean()
```

```python
p = df2.sort_index()
wakebat_mean = p.to_frame().T
wakebat_mean
```

```python
# Iterate through each row in the DataFrame
for index, row in result.iterrows():
    # Divide each entry by the corresponding 'G' column entry
    for column in result.columns:
        result.at[index, column] = row[column] / row['G']
        
result.head()
```
