```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Check for missing values
missing_values = df.isnull().sum().sort_values(ascending=False)
print("Missing values per column:")
print(missing_values)

# Visualize missing values using a heatmap
plt.figure(figsize=(15, 8))
sns.heatmap(df.isnull(), cbar=False, cmap='viridis')
plt.title('Missing Values in the Dataset')
plt.show()

# Drop columns with more than 50% missing values
cols_to_drop = missing_values[missing_values > len(df) * 0.5].index
df.drop(columns=cols_to_drop, inplace=True)
print(f"Dropped columns: {list(cols_to_drop)}")

# Fill missing categorical values with the mode and numerical with the median
for col in df.select_dtypes(include=['object']).columns:
    df[col].fillna(df[col].mode()[0], inplace=True)

for col in df.select_dtypes(include=['float64', 'int64']).columns:
    df[col].fillna(df[col].median(), inplace=True)

# Convert boolean-like columns (e.g., 'yes', 'no') to True/False
bool_columns = ['arePetsAllowed', 'hasFlatSharingCommunity', 'isUnderRoof',
                'isNewBuilding', 'hasGarage', 'hasParking', 'isOldBuilding',
                'hasSwimmingPool', 'hasFireplace', 'isCornerHouse',
                'isMiddleHouse', 'hasGardenShed', 'hasLiftingPlatform',
                'hasBalcony', 'hasCableTv', 'isGroundFloorRaised',
                'hasNiceView', 'hasConnectedBuildingLand', 'hasElevator',
                'isChildFriendly']

for col in bool_columns:
    if col in df.columns:
        df[col] = df[col].map({'yes': True, 'no': False})


```

```python
# Check for any remaining missing values
print("Remaining missing values per column:")
print(df.isnull().sum().sort_values(ascending=False))

```

```python
# EDA: Distribution of property types
plt.figure(figsize=(10, 6))
sns.countplot(x='type', data=df, order=df['type'].value_counts().index)
plt.xticks(rotation=45)
plt.title('Distribution of Property Types')
plt.show()
```

```python

# EDA: Distribution of rentNet
plt.figure(figsize=(10, 6))
sns.histplot(df['rentNet'], kde=True, bins=30, color='skyblue')
plt.title('Distribution of Net Rent')
plt.xlabel('Net Rent')
plt.ylabel('Frequency')
plt.show()


```

```python
# EDA: Living space vs. rentNet scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x='livingSpace', y='rentNet', hue='type', data=df)
plt.title('Living Space vs. Net Rent')
plt.xlabel('Living Space (m²)')
plt.ylabel('Net Rent (CHF)')
plt.show()

```

```python
# EDA: Correlation heatmap for numerical features
plt.figure(figsize=(15, 10))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap of Numerical Features')
plt.show()
```

```python

import pandas as pd
import folium



# Create a base map centered around the average latitude and longitude
latitude = df['lat'].mean()
longitude = df['lon'].mean()
base_map = folium.Map(location=[latitude, longitude], zoom_start=12)

# Function to add markers to the map
def add_markers(map_object, data):
    for _, row in data.iterrows():
        # Customize the popup content to display key information
        popup_text = (f"ID: {row['ID']}<br>"
                      f"Type: {row['type']}<br>"
                      f"City: {row['city']}<br>"
                      f"Address: {row['address']}<br>"
                      f"Net Rent: {row['rentNet']} {row['currency']}<br>"
                      f"Living Space: {row['livingSpace']} m²<br>"
                      f"Rooms: {row['numberOfRooms']}")
        
        # Add a marker for each property
        folium.Marker(
            location=[row['lat'], row['lon']],
            popup=folium.Popup(popup_text, max_width=300),
            icon=folium.Icon(color='blue', icon='home')
        ).add_to(map_object)

# Add the property markers to the base map
add_markers(base_map, df)

# Display the map
base_map
```

```python
# Assuming `average_chf_per_postcode` has already been created and contains 'postcode' and 'CHF/m2'
# Step 1: Rename 'CHF/m2' in `average_chf_per_postcode` to distinguish it as the average
average_chf_per_postcode = average_chf_per_postcode.rename(columns={'CHF/m2': 'average_CHF_per_postcode'})

# Step 2: Merge `sorted_df` with `average_chf_per_postcode` on the 'postcode' column
sorted_df = pd.merge(sorted_df, average_chf_per_postcode[['postcode', 'average_CHF_per_postcode']], on='postcode', how='left')

# Step 3: Verify the updated DataFrame
sorted_df.head()
```

```python
# Import additional libraries for color mapping
import folium
import geopandas as gpd
from matplotlib import cm, colors

# Assuming `sorted_df` already contains 'lat' and 'lon' columns
# Calculate the central point for the map
latitude = sorted_df['lat'].mean()
longitude = sorted_df['lon'].mean()

# Create a base folium map centered around the average latitude and longitude
map_object = folium.Map(location=[latitude, longitude], zoom_start=10)

# Set up a color map using matplotlib (e.g., a viridis colormap)
colormap = cm.get_cmap('viridis')
normalize = colors.Normalize(vmin=sorted_df['CHF/m2'].min(), vmax=sorted_df['CHF/m2'].max())

# Add CircleMarkers with dynamic coloring based on CHF/m2
for _, row in sorted_df.iterrows():
    # Map the CHF/m2 value to a color
    color_value = colormap(normalize(row['CHF/m2']))
    hex_color = colors.rgb2hex(color_value)  # Convert the RGB color to HEX for folium

    folium.CircleMarker(
        location=[row['lat'], row['lon']],
        radius=8,
        fill=True,
        fill_color=hex_color,  # Use the dynamically generated color
        color='grey',
        fill_opacity=0.7,
        popup=f"Postcode: {row['postcode']}<br>CHF/m2: {row['CHF/m2']}"
    ).add_to(map_object)

# Add a legend to the map for reference
colormap_legend = folium.LinearColormap(
    [colors.rgb2hex(colormap(normalize(val))) for val in sorted_df['CHF/m2'].quantile([0, 0.25, 0.5, 0.75, 1])],
    vmin=sorted_df['CHF/m2'].min(),
    vmax=sorted_df['CHF/m2'].max(),
    caption='CHF/m2 by Postcode'
)
colormap_legend.add_to(map_object)

# Display the map
map_object
```
