```python
import pandas as pd
import plotly.express as px

# Ensure the 'start_date' is in datetime format
df['start_date'] = pd.to_datetime(df['start_date'], format='%m/%d/%y')

# Filter data to include only 'Kamala Harris' and 'Donald Trump' from August 2024 onwards
df_filtered = df[(df['candidate_name'].isin(['Kamala Harris', 'Donald Trump'])) & 
                 (df['start_date'] >= '2024-08-01')]

# Create an interactive line plot using Plotly Express
fig = px.scatter(
    df_filtered, 
    x='start_date', 
    y='pct', 
    color='candidate_name',
    trendline='lowess',  # Add a smooth trendline (LOWESS)
    labels={'pct': 'Polling Percentage (%)', 'start_date': 'Date'},
    title='Polling Percentage Over Time (Aug 2024 Onwards)',
    hover_data=['display_name', 'sample_size']
)

# Customize layout
fig.update_layout(
    title='Who’s Ahead in the National Polls?',
    title_x=0.5,  # Center title
    yaxis=dict(title='Polling Percentage (%)', range=[0, 100]),
    xaxis=dict(title='Date', tickformat='%b %d, %Y'),  # Format date axis
    legend=dict(title='Candidate', orientation='h', yanchor='bottom', y=1.02, xanchor='right', x=1)
)


# Display the figure
fig.show()
```

```python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go

# Filter the data to only include Kamala Harris and Donald Trump
df_filtered = df_clean[df_clean['candidate_name'].isin(['Kamala Harris', 'Donald Trump'])]

# Create the scatter plot with larger markers and reduced opacity
fig = px.scatter(
    df_filtered,
    x='start_date',
    y='pct',
    color='candidate_name',
    color_discrete_map={'Kamala Harris': 'blue', 'Donald Trump': 'red'},
    labels={'start_date': 'Date', 'pct': 'Polling Percentage (%)'},
    title="Who's Ahead in the National Polls?",
    opacity=0.7,  # Reducing opacity for better visibility with overlapping points
    size_max=8  # Increasing marker size
)

# Add smoothed trendlines using rolling mean
for candidate in ['Kamala Harris', 'Donald Trump']:
    candidate_data = df_filtered[df_filtered['candidate_name'] == candidate]
    fig.add_trace(
        go.Scatter(
            x=candidate_data['start_date'],
            y=candidate_data['pct'].rolling(window=7).mean(),  # 7-day smoothing
            mode='lines',
            name=f"{candidate} (Trend)",
            line=dict(
                color='blue' if candidate == 'Kamala Harris' else 'red',
                width=3  # Slightly thicker lines for visibility
            )
        )
    )

# Customize the layout
fig.update_layout(
    plot_bgcolor='white',
    xaxis=dict(
        showgrid=True, gridcolor='lightgray',
        title=dict(text='Date'),
        tickformat='%b %d, %Y',  # Formatting x-axis dates
        tickangle=45  # Rotating x-axis labels for readability
    ),
    yaxis=dict(
        showgrid=True, gridcolor='lightgray',
        range=[30, 60],  # Expanding y-axis range slightly for better spacing
        title=dict(text='Polling Percentage (%)')
    ),
    title=dict(
        text="Who's Ahead in the National Polls?",
        font=dict(size=24)  # Larger title for better visibility
    ),
    legend=dict(
        title='Candidate',
        orientation='h',
        yanchor='bottom',
        y=1.02,
        xanchor='right',
        x=1
    )
)

# Show the interactive figure
fig.show()
```
