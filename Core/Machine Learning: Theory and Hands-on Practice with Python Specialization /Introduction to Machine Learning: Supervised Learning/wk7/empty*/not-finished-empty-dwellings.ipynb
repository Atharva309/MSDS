```python
import pandas as pd
import os
import warnings

# Suppress warnings for cleaner output
warnings.simplefilter(action='ignore', category=FutureWarning)

standard_headers = [
    'location', 'total dwelling stock', '1 Room', '2 Rooms', '3 Rooms',
    '4 Rooms', '5 Rooms', '6+ Rooms', 'Total', 'of which are SFH',
    'of which are new', 'of which are for rent', 'of which are for sale',
    'Proportion of empty dwelling'
]

def process_sheet(sheet_df, standard_headers):
    try:
        # Ensure the first column is converted to string where possible
        sheet_df.iloc[:, 0] = sheet_df.iloc[:, 0].astype(str)

        # Apply string operations safely, skipping non-string values
        valid_strings = pd.Series(sheet_df.iloc[:, 0], dtype="string")
        total_row_index = valid_strings[valid_strings.str.contains('Total', case=False, na=False)].index

        if len(total_row_index) == 0:
            print("No 'Total' row found.")
            return None

        # Use the identified row as the header
        header_row = total_row_index[0]
        sheet_clean = sheet_df.iloc[header_row:].copy()
        sheet_clean.columns = sheet_clean.iloc[0]  # Set new headers
        sheet_clean = sheet_clean[1:].reset_index(drop=True)

        # Clean column names
        sheet_clean.columns = sheet_clean.columns.str.strip()

        # Rename columns according to the standardized headers
        rename_dict = {sheet_clean.columns[i]: col for i, col in enumerate(standard_headers) if i < len(sheet_clean.columns)}
        sheet_clean.rename(columns=rename_dict, inplace=True)

        # Filter only the relevant columns
        existing_headers = [col for col in standard_headers if col in sheet_clean.columns]
        sheet_clean = sheet_clean[existing_headers]

        # Clean and convert numeric columns
        numeric_columns = [col for col in standard_headers[1:] if col in sheet_clean.columns]
        for col in numeric_columns:
            sheet_clean[col] = pd.to_numeric(sheet_clean[col].str.replace(',', '', regex=True), errors='coerce')

        # Drop rows where all numeric columns are NaN
        sheet_clean.dropna(subset=numeric_columns, how='all', inplace=True)

        # Reset index for a clean output
        sheet_clean.reset_index(drop=True, inplace=True)
        return sheet_clean

    except Exception as e:
        print(f"Error processing sheet: {e}")
        return None

def process_all_sheets(file_path, standard_headers):
    if not os.path.exists(file_path):
        raise FileNotFoundError(f"The file '{file_path}' does not exist.")

    try:
        excel = pd.ExcelFile(file_path, engine='openpyxl')
    except Exception as e:
        print(f"Error loading Excel file: {e}")
        return None

    sheet_names = excel.sheet_names
    print(f"Sheets found: {sheet_names}\n")

    processed_dfs = {}
    for sheet in sheet_names:
        print(f"Processing sheet: {sheet}")
        try:
            df_raw = pd.read_excel(file_path, sheet_name=sheet, header=None, dtype=str)
            df_clean = process_sheet(df_raw, standard_headers)
            if df_clean is not None:
                processed_dfs[sheet] = df_clean
                print(f"Successfully processed sheet: {sheet}\n")
            else:
                print(f"Skipped sheet: {sheet}.\n")
        except Exception as e:
            print(f"Error processing sheet '{sheet}': {e}\n")

    print("All sheets have been processed.\n")
    return processed_dfs

# Execution
file_path = 'Empty dwellings by canton.xls'
dfs = process_all_sheets(file_path, standard_headers)

# Preview results
if dfs:
    for sheet_name, df in dfs.items():
        print(f"Preview of '{sheet_name}':")
        print(df.head(), "\n")
```

```python
import pandas as pd
import os
import warnings

# Suppress warnings for cleaner output
warnings.simplefilter(action='ignore', category=FutureWarning)

# Define the desired standardized column headers
standard_headers = [
    'location',
    'total dwelling stock',
    '1 Room',
    '2 Rooms',
    '3 Rooms',
    '4 Rooms',
    '5 Rooms',
    '6+ Rooms',
    'Total',
    'of which are SFH',
    'of which are new',
    'of which are for rent',
    'of which are for sale',
    'Proportion of empty dwelling'
]

def process_sheet(sheet_df, standard_headers):
    """
    Processes a single sheet DataFrame to set standardized headers and clean data.

    Parameters:
    sheet_df (DataFrame): The raw DataFrame of the sheet.
    standard_headers (list): The list of standardized column headers.

    Returns:
    DataFrame: Cleaned DataFrame with standardized headers.
    """
    # Identify the row index where the first column contains 'Total'
    total_row_index = sheet_df[sheet_df.iloc[:, 0].astype(str).str.contains('Total', case=False, na=False)].index

    if len(total_row_index) == 0:
        print("No 'Total' row found in this sheet.")
        return None

    # Assume the first occurrence is the header row
    header_row = total_row_index[0]

    # Extract the data starting from the header row
    sheet_clean = sheet_df.iloc[header_row:].copy()

    # Set the header using the identified row
    sheet_clean.columns = sheet_clean.iloc[0]

    # Remove the header row from data
    sheet_clean = sheet_clean[1:]

    # Reset index for cleanliness
    sheet_clean.reset_index(drop=True, inplace=True)

    # Trim whitespace from column names
    sheet_clean.columns = sheet_clean.columns.str.strip()

    # Rename columns to standardized headers
    # First, identify current columns
    current_columns = sheet_clean.columns.tolist()

    # Create a mapping dictionary only for columns that exist
    rename_dict = {}
    for idx, new_col in enumerate(standard_headers):
        if idx < len(current_columns):
            old_col = current_columns[idx]
            rename_dict[old_col] = new_col

    # Rename the columns
    sheet_clean.rename(columns=rename_dict, inplace=True)

    # Select only the standardized columns that exist in the current sheet
    existing_standard_headers = [col for col in standard_headers if col in sheet_clean.columns]
    sheet_clean = sheet_clean[existing_standard_headers]

    # Clean data: Remove any rows that do not have valid location names
    # Assuming location names are strings and other columns are numeric
    if 'location' in sheet_clean.columns:
        sheet_clean = sheet_clean[sheet_clean['location'].notna()]
    else:
        print("No 'location' column found after renaming.")
        return None

    # Optionally, remove any footer notes or irrelevant rows
    # For example, remove rows where 'location' contains 'Basis' or similar keywords
    footer_keywords = ['Basis', 'Information', '©', 'Data as of']
    for keyword in footer_keywords:
        sheet_clean = sheet_clean[~sheet_clean['location'].astype(str).str.contains(keyword, case=False, na=False)]

    # Replace any commas in numeric columns and convert to appropriate data types
    numeric_columns = [col for col in standard_headers[1:-1] if col in sheet_clean.columns]
    for col in numeric_columns:
        sheet_clean[col] = sheet_clean[col].astype(str).str.replace(',', '').str.replace('−', '-').str.strip()
        # Handle empty strings or non-numeric entries
        sheet_clean[col] = pd.to_numeric(sheet_clean[col], errors='coerce')

    # Convert 'Proportion of empty dwelling' to float if it exists
    proportion_col = 'Proportion of empty dwelling'
    if proportion_col in sheet_clean.columns:
        sheet_clean[proportion_col] = sheet_clean[proportion_col].astype(str).str.replace(',', '').str.strip()
        sheet_clean[proportion_col] = pd.to_numeric(sheet_clean[proportion_col], errors='coerce')

    # Drop rows with all NaN in numeric columns (if any)
    drop_cols = numeric_columns.copy()
    if proportion_col in sheet_clean.columns:
        drop_cols.append(proportion_col)
    sheet_clean.dropna(subset=drop_cols, how='all', inplace=True)

    # Reset index after cleaning
    sheet_clean.reset_index(drop=True, inplace=True)

    return sheet_clean

def process_all_sheets(excel_file_path, standard_headers):
    """
    Processes all sheets in the given Excel workbook.

    Parameters:
    excel_file_path (str): The path to the Excel workbook.
    standard_headers (list): The list of standardized column headers.

    Returns:
    dict: Dictionary of cleaned DataFrames with sheet names as keys.
    """
    # Check if the file exists
    if not os.path.exists(excel_file_path):
        raise FileNotFoundError(f"The file '{excel_file_path}' does not exist.")

    # Load the Excel file with xlrd engine for .xls files
    try:
        excel = pd.ExcelFile(excel_file_path, engine='xlrd')
    except Exception as e:
        print(f"Error loading Excel file: {e}")
        return None

    # Get all sheet names
    sheet_names = excel.sheet_names

    print(f"Sheets found: {sheet_names}\n")

    # Initialize a dictionary to hold processed DataFrames
    processed_dfs = {}

    # Loop through each sheet and process
    for sheet in sheet_names:
        print(f"Processing sheet: {sheet}")

        try:
            # Read the sheet without setting a header
            df_raw = pd.read_excel(excel, sheet_name=sheet, header=None, dtype=str)

            # Process the sheet
            df_clean = process_sheet(df_raw, standard_headers)

            if df_clean is not None:
                # Assign the cleaned DataFrame to the dictionary with sheet name as key
                processed_dfs[sheet] = df_clean
                print(f"Successfully processed sheet: {sheet}\n")
            else:
                print(f"Skipped sheet: {sheet} due to missing 'Total' row or 'location' column.\n")

        except Exception as e:
            print(f"Error processing sheet '{sheet}': {e}\n")
            continue

    print("All sheets have been processed.\n")
    return processed_dfs

# --- Execution Starts Here ---

# Specify the path to your Excel workbook
excel_file_path = 'Empty dwellings by canton.xls'  # Replace with your actual file path

# Process all sheets and get the dictionary of DataFrames
dfs = process_all_sheets(excel_file_path, standard_headers)

# Example: Accessing a specific DataFrame
# df_2024 = dfs.get('2024 (basis BDS2023)')

# To view the processed DataFrames, you can iterate through the dictionary
# For example, to display the first few rows of each DataFrame:
for sheet_name, df in dfs.items():
    print(f"Preview of '{sheet_name}':")
    print(df.head(), "\n")
```

```python
import pandas as pd
import os
import warnings

# Suppress warnings for cleaner output
warnings.simplefilter(action='ignore', category=FutureWarning)

# Define the desired standardized column headers
standard_headers = [
    'location',
    'total dwelling stock',
    '1 Room',
    '2 Rooms',
    '3 Rooms',
    '4 Rooms',
    '5 Rooms',
    '6+ Rooms',
    'Total',
    'of which are SFH',
    'of which are new',
    'of which are for rent',
    'of which are for sale',
    'Proportion of empty dwelling'
]

def process_specific_sheet(excel_file_path, sheet_name, standard_headers):
    """
    Processes a specific sheet in the Excel workbook to set standardized headers and clean data.

    Parameters:
    excel_file_path (str): The path to the Excel workbook.
    sheet_name (str): The name of the sheet to process.
    standard_headers (list): The list of standardized column headers.

    Returns:
    DataFrame: Cleaned DataFrame with standardized headers.
    """
    # Check if the file exists
    if not os.path.exists(excel_file_path):
        raise FileNotFoundError(f"The file '{excel_file_path}' does not exist.")

    try:
        # Read the entire sheet without setting a header
        df_raw = pd.read_excel(excel_file_path, sheet_name=sheet_name, header=None, dtype=str)
    except Exception as e:
        print(f"Error loading sheet '{sheet_name}': {e}")
        return None

    # Identify header rows
    # Assuming headers are in the first 6 rows, adjust if necessary
    potential_header_rows = df_raw.head(6)

    # Find the row that contains 'Total dwelling stock' to identify header start
    header_start_candidates = potential_header_rows[
        potential_header_rows.iloc[:, 0].astype(str).str.contains('Total dwelling stock', case=False, na=False)
    ].index.tolist()

    if not header_start_candidates:
        print(f"No 'Total dwelling stock' row found in sheet '{sheet_name}'.")
        return None

    header_start = header_start_candidates[0]
    num_header_rows = 4  # Based on the provided sample, adjust if necessary

    # Ensure we don't go out of bounds
    if header_start + num_header_rows > len(df_raw):
        print(f"Not enough rows for headers in sheet '{sheet_name}'.")
        return None

    # Extract header rows
    header_rows = df_raw.iloc[header_start:header_start + num_header_rows]

    # Combine header rows to form multi-level column names
    combined_headers = header_rows.fillna('').astype(str).agg(' '.join, axis=0).str.replace('\n', ' ').str.strip()

    # Assign combined headers to the data
    data_start = header_start + num_header_rows
    df_clean = df_raw.iloc[data_start:].copy()

    # Set the combined headers
    df_clean.columns = combined_headers

    # Trim whitespace from column names
    df_clean.columns = df_clean.columns.str.strip()

    # Rename columns to standardized headers
    current_columns = df_clean.columns.tolist()

    # Create a mapping dictionary only for columns that exist
    rename_dict = {}
    for idx, new_col in enumerate(standard_headers):
        if idx < len(current_columns):
            old_col = current_columns[idx]
            rename_dict[old_col] = new_col

    # Rename the columns
    df_clean.rename(columns=rename_dict, inplace=True)

    # Select only the standardized columns that exist in the current sheet
    existing_standard_headers = [col for col in standard_headers if col in df_clean.columns]
    df_clean = df_clean[existing_standard_headers]

    # Remove the 'Total' row if present
    if 'location' in df_clean.columns:
        df_clean = df_clean[df_clean['location'].astype(str).str.lower() != 'total']
    else:
        print(f"No 'location' column found after renaming in sheet '{sheet_name}'.")
        return None

    # Remove any footer notes or irrelevant rows based on keywords
    footer_keywords = ['Basis', 'Information', '©', 'Data as of']
    for keyword in footer_keywords:
        df_clean = df_clean[~df_clean['location'].astype(str).str.contains(keyword, case=False, na=False)]

    # Replace any commas and handle negative signs in numeric columns, then convert to appropriate data types
    numeric_columns = [col for col in standard_headers[1:-1] if col in df_clean.columns]
    for col in numeric_columns:
        df_clean[col] = df_clean[col].astype(str).str.replace(',', '').str.replace('−', '-').str.strip()
        # Handle empty strings or non-numeric entries
        df_clean[col] = pd.to_numeric(df_clean[col], errors='coerce')

    # Convert 'Proportion of empty dwelling' to float if it exists
    proportion_col = 'Proportion of empty dwelling'
    if proportion_col in df_clean.columns:
        df_clean[proportion_col] = df_clean[proportion_col].astype(str).str.replace(',', '').str.strip()
        df_clean[proportion_col] = pd.to_numeric(df_clean[proportion_col], errors='coerce')

    # Drop rows with all NaN in numeric columns (if any)
    drop_cols = numeric_columns.copy()
    if proportion_col in df_clean.columns:
        drop_cols.append(proportion_col)
    df_clean.dropna(subset=drop_cols, how='all', inplace=True)

    # Reset index after cleaning
    df_clean.reset_index(drop=True, inplace=True)

    return df_clean

# --- Execution Starts Here ---

# Specify the path to your Excel workbook
excel_file_path = 'Empty dwellings by canton.xls'  # Replace with your actual file path

# Specify the sheets you want to process
target_sheets = [
    '2022 (basis BDS2021)',
    '2021 (basis BDS2020)',
    '2020 (basis BDS2019)',
    '2019 (basis BDS2018)',
    '2018 (basis BDS2017)',
    '2017 (basis BDS2016)',
    '2016 (basis BDS2015)',
    '2015 (basis BDS2014)',
    '2014 (basis BDS2013)',
    '2013 (basis BDS2012)',
    '2012 (basis BDS2011)',
    '2011 (new basis BDS2010)',
    '2010 (new basis BDS2009)',
    '2010',
    '2009',
    '2008'
]

# Initialize a dictionary to hold processed DataFrames
processed_dfs = {}

# Loop through each target sheet and process
for sheet in target_sheets:
    print(f"Processing sheet: {sheet}")
    df = process_specific_sheet(excel_file_path, sheet, standard_headers)
    if df is not None:
        processed_dfs[sheet] = df
        print(f"Successfully processed sheet: {sheet}\n")
    else:
        print(f"Failed to process sheet: {sheet}\n")

print("All targeted sheets have been processed.\n")

# Example: Accessing and displaying the processed '2022 (basis BDS2021)' DataFrame
df_2022 = processed_dfs.get('2022 (basis BDS2021)')
if df_2022 is not None:
    print("df_2022:")
    print(df_2022)
else:
    print("df_2022 is None.")
```
