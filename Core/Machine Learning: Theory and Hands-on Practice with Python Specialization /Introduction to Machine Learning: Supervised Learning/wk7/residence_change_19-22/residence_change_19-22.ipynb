```python
import pandas as pd

# Define the path to your Excel file
file_path = 'Annual-change-in-the-number-of-residences-by-type-of-activity-and-by-canton.xlsx'

# Read all sheets into a dictionary of DataFrames
all_sheets = pd.read_excel(file_path, sheet_name=None, skiprows=2, engine='openpyxl')

# Initialize an empty dictionary to hold cleaned DataFrames
cleaned_sheets = {}

# Loop through each sheet and apply the same cleaning process
for sheet_name, df in all_sheets.items():
    # Drop the first row (index 0) containing duplicate headers
    df = df.drop(index=0)

    # Rename columns to clean up the header names
    df.columns = [
        'Municipality', 'Dwellings', 'New Buildings', 'Conversions',
        'Demolition', 'Conversion Losses', 'Net Additional Dwellings', 'Corrections'
    ]

    # Convert 'Municipality' to string for consistency
    df['Municipality'] = df['Municipality'].astype(str)

    # Store the cleaned DataFrame in the dictionary
    cleaned_sheets[sheet_name] = df

# Display the names of the cleaned sheets and preview one of them
print("Cleaned Sheets:", cleaned_sheets.keys())
print("Preview of '2022' sheet after cleaning:")
print(cleaned_sheets['2022'].head())
```
