```python
import pandas as pd

# Specify the path to your Excel file
file_path = "Empty dwellings by canton.xlsx"

# Read all sheets into a dictionary of DataFrames
dfs = pd.read_excel(file_path, sheet_name=None, header=[3, 4, 5], skiprows=6)

# Create an empty dictionary to store cleaned DataFrames for each sheet
cleaned_dfs = {}

# Iterate through each sheet and apply the cleaning steps
for sheet_name, df in dfs.items():
    # Print the current sheet being processed
    print(f"Processing sheet: {sheet_name}")



    # Drop rows with any NaN values
    df_cleaned = df.dropna(how='any')

    # Store the cleaned DataFrame in the dictionary
    cleaned_dfs[sheet_name] = df_cleaned

    # Optional: Display the cleaned DataFrame for verification
    print(f"Cleaned DataFrame for {sheet_name}:\n{df_cleaned.head()}\n")

# At this point, `cleaned_dfs` is a dictionary containing the cleaned DataFrames for each sheet
```
