### Question 1

Twelve fish are randomly sampled from a salmon hatchery and their lengths are measured. The average length of fish in this sample is $82.17$ cm.

Assume that the true distribution of fish lengths is normally distributed with standard deviation $\sigma = 2.2$. 

Test:

$$H_0: \mu = 81 \quad \text{versus} \quad H_1: \mu \neq 81$$

using a $0.05$ level of significance.

What is your test statistic, critical value(s), and conclusion?

**Answer:**

$1.8423, \pm 1.96, \text{Fail to reject } H_0$


### Question 2

Let $\gamma(\theta)$ be the power function for a hypothesis test, with level of significance $\alpha$, involving a parameter $\theta$.

Which of the following are necessarily true? 

(Check all that apply.)

- $\gamma(\theta)$ can only take values between 0 and 1 (inclusive).


  For any fixed $\theta$, $\gamma(\theta)$ is a probability. It is the probability of rejecting $H_0$ when the true value of the parameter is $\theta$.

- The maximum value of $\gamma(\theta)$ for all $\theta$ in $H_0$ is $\alpha$.

    $\alpha$ is the maximum probability of making a Type I error, which is the maximum probability of rejecting $H_0$ when we should not because $H_0$ is true.

### Question 4

Is there evidence that the true mean price of Frosted Toasty Bites in the US is less than 4.69?

Test the relevant hypotheses at the 0.10 level of significance.

**Answer:**

### Question 8

A random sample of size 33 from the $N(\mu_1, \sigma_1^2)$ distribution has a sample variance of $s_1^2 = 7.8$.

An independent random sample of size 55 from a $N(\mu_2, \sigma_2^2)$ distribution has a sample variance of $s_2^2 = 6.3$.

Is there evidence to suggest that $\sigma_2^2 < \sigma_1^2$?

Consider testing the hypotheses:

$$H_0: \sigma_2^2 = \sigma_1^2 \quad \text{versus} \quad H_1: \sigma_2^2 < \sigma_1^2$$

using $\alpha = 0.05$ level of significance and the test statistic $\frac{S_2^2}{S_1^2}$. What is the appropriate critical value to use for this test? Give your answer to 3 decimal places.

(Note: This is not the $c$ in a rejection rule like "Reject $H_0$, in favor of $H_1$ if $\frac{S_2^2}{S_1^2} < c$" or "Reject $H_0$, in favor of $H_1$ if $\frac{S_2^2}{S_1^2} > c$." It is the critical value that goes into solving for $c$.)

**Answer:**

The appropriate critical value for this test is $0.144$.
