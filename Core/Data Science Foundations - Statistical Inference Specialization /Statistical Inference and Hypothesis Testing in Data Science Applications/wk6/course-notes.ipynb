Suppose that $X_1, X_2, \dots, X_n$ is a random sample from the $N(\mu, \sigma^2)$ distribution where $\sigma^2$ is known.
Let’s construct a test of size (level of significance) $\alpha$ for

$$H_0 : \mu = \mu_0 \quad \text{versus} \quad H_1 : \mu = \mu_1$$

where $\mu_0$ and $\mu_1$ are fixed and known.

**Answer:**

The test when $\mu_0 > \mu_1$ is to

"Reject $H_0$, in favor of $H_1$ if $\bar{X} < \mu_0 - z_\alpha \frac{\sigma}{\sqrt{n}}$."

A random sample of 500 people in a certain country, which is about to have a national election, were asked whether they preferred “Candidate A” or “Candidate B”.
From this sample, 320 people responded that they preferred Candidate A.
Let $p$ be the true proportion of the people in the country who prefer Candidate A.

Test the hypotheses:

$$H_0 : p \leq 0.65 \quad \text{versus} \quad H_1 : p > 0.65$$

Use a level of significance of 0.10.

We have an estimate $\hat{p} = \frac{320}{500} = \frac{16}{25}$.

**Answer:**

Reject $H_0$ if
$$
\hat{p} > 0.65 + z_{0.10} \sqrt{\frac{0.65(1 - 0.65)}{n}}.
$$

We fail to reject $H_0$ in favor of $H_1$.

The data do not suggest that the true proportion of people who like Candidate A is greater than 0.65.

Derive a test to figure out whether or not the two population means are the same.

$$H_0 : \mu_1 = \mu_2$$
$$H_1 : \mu_1 \neq \mu_2$$

**Answer:**

$$
|\bar{X}_1 - \bar{X}_2| > z_{\alpha/2} \sqrt{\frac{\sigma_1^2}{n_1} + \frac{\sigma_2^2}{n_2}},
$$
we reject $H_0$. The data suggests that the true mean scores for the counties are different!

### Wk 4


Suppose that $X_1, X_2, \dots, X_n$ is a random sample from the exponential distribution with rate $\lambda > 0$.
Derive a hypothesis test of size $\alpha$ for

$$H_0 : \lambda = \lambda_0 \quad \text{vs.} \quad H_1 : \lambda > \lambda_0$$

What statistic should we use?

**Answer:**

**Test 1:**

Reject $H_0$, in favor of $H_1$, if
$$
\sum_{i=1}^{n} X_i < \frac{\chi^2_{1-\alpha, 2n}}{2n\lambda_0}
$$

where $\chi^2_{\alpha, n}$ is the critical value from the chi-squared distribution with $n$ degrees of freedom.

**Test 2:**

Reject $H_0$, in favor of $H_1$, if
$$
Y_n = \min(X_1, X_2, \dots, X_n) < -\frac{\ln(1 - \alpha)}{n\lambda_0}
$$

**Compare the Tests**

$$H_0 : \lambda = 1 \quad \text{vs.} \quad H_1 : \lambda > 1$$


---



Suppose that $X_1, X_2, \dots, X_n$ is a random sample from the exponential distribution with rate $\lambda > 0$.

Derive a uniformly most powerful hypothesis test of size $\alpha$ for

$$H_0 : \lambda = \lambda_0 \quad \text{vs.} \quad H_1 : \lambda > \lambda_0$$

**Answer:**

The uniformly most powerful test for $\lambda_1 > \lambda_0$ is to reject $H_0$, in favor of $H_1$, if
$$
\sum_{i=1}^{n} X_i < \frac{\chi^2_{1-\alpha, 2n}}{2n\lambda_0}.
$$

Note that this test does not depend on the particular value of $\lambda_1$ as long as $\lambda_1 > \lambda_0$.

The best (most powerful) test of
$$H_0 : \lambda = \lambda_0 \quad \text{vs.} \quad H_1 : \lambda = \lambda_1$$
is the uniformly most powerful (best) test for
$$H_0 : \lambda = \lambda_0 \quad \text{vs.} \quad H_1 : \lambda > \lambda_0$$

**Does there exist a “UMP” test for $\lambda_1 \neq \lambda_0$?**

**No**

There is no one best test that we can use for all $\lambda_1 \neq \lambda_0$.
