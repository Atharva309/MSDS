5. Consider the distribution with pdf
$$f(x; \theta) = 1 - \theta^2\left(x - \frac{1}{2}\right), \quad 0 < x < 1, \quad -1 < \theta < 1.$$
  
  a. Find the best test of size $\alpha$ for
  $$H_0 : \theta = 0 \quad \text{versus} \quad H_1 : \theta = \theta_1,$$
  for some fixed $\theta_1 \neq 0$, based on a sample of size 1.
  
  b. Find (if it exists) a UMP (uniformly most powerful) test of size $\alpha$ for
  $$H_0 : \theta = 0 \quad \text{versus} \quad H_1 : \theta \neq 0$$
  based on a sample of size 1.

**Answer:**

a.
This results in the rejection region:
Reject $H_0$ if $x < \frac{1}{2} - c$ or $x > \frac{1}{2} + c$,

where $c$ is such that the probability of rejection under $H_0$ is exactly $\alpha$.

b.
This two-sided test would reject $H_0$ if the observed value $x$ is sufficiently far from $\frac{1}{2}$, similar to the setup in part (a).

6. Let $X_1, X_2, \dots, X_n$ be a random sample of size $n$ from the $N(0, \sigma^2)$ distribution. Derive the UMP test of size $\alpha$ for 
$$H_0 : \sigma^2 = \sigma^2_0 \quad \text{versus} \quad H_1 : \sigma^2 > \sigma^2_0.$$

**Answer:**

Reject $H_0$ if:
$$
\frac{nS^2}{\sigma_0^2} > \chi^2_{n, 1-\alpha},
$$

where $\chi^2_{n, 1-\alpha}$ is the critical value from the chi-squared distribution with $n$ degrees of freedom corresponding to the significance level $\alpha$.

9. Let $X_1, X_2, \dots, X_n$ be a random sample from the exponential distribution with rate $\theta$. Consider the hypotheses 
$$H_0 : \theta = \theta_0 \quad \text{and} \quad H_1 : \theta < \theta_0.$$
  
  a. Find a test of size $\alpha$ based on the minimum of the sample.
  
  b. Find the UMP test of size $\alpha$.
  
  c. Find and compare the power functions of your two tests.

**Answer:**

a. Reject $H_0: \theta = \theta_0$ in favor of $H_1: \theta < \theta_0$ if:
$$
X_{(1)} \leq -\frac{\log(1-\alpha)}{n\theta_0}.
$$

b. The test derived in part (a) is indeed the UMP test.

c. The power function of the UMP test is:
$$
\beta(\theta) = 1 - (1-\alpha)^{\theta/\theta_0}.
$$

10. Consider a random sample of size $n$ from the $\text{Unif}(0, \theta)$ distribution. Find the UMP test of size $\alpha$ for 
$$H_0 : \theta \geq \theta_0 \quad \text{versus} \quad H_1 : \theta < \theta_0.$$


**Answer:**

Reject $H_0$ if:
$$
X_{(n)} \leq \alpha^{1/n} \cdot \theta_0.
$$

This test is uniformly most powerful because it maximizes the probability of rejecting $H_0$ for any $\theta < \theta_0$ while maintaining the size $\alpha$.

12. Consider a random sample $X_1, X_2, \dots, X_n$ from a distribution with pdf 
$$f(x; \theta) = \theta(1 - x)^{\theta-1}, \quad 0 < x < 1, \quad \theta > 0.$$
Give the form of the GLRT (generalized likelihood ratio test) for testing 
$$H_0 : \theta = 1 \quad \text{against} \quad H_1 : \theta \neq 1.$$

**Answer:**

Reject $H_0$ if:
$$
-2\log \Lambda = 2n\log \hat{\theta} + 2(\hat{\theta} - 1)\sum_{i=1}^{n}\log(1 - X_i)
$$

is greater than a critical value $c$, where $\hat{\theta}$ is the MLE of $\theta$ given by

$$
\hat{\theta} = -\frac{n}{\sum_{i=1}^{n}\log(1 - X_i)}.
$$

14. Let $X_1, X_2, \dots, X_n$ be a random sample from the $N(\mu, \sigma^2)$ distribution where $\sigma^2$ is known. Derive the GLRT of size $\alpha$ for testing 
$$H_0 : \mu = \mu_0 \quad \text{versus} \quad H_1 : \mu \neq \mu_0.$$

**Answer:**


Reject $H_0$ if:
$$
\frac{n(\bar{X} - \mu_0)^2}{\sigma^2} > \chi^2_{1,\alpha}.
$$

This test compares the squared deviation of the sample mean $\bar{X}$ from $\mu_0$ (normalized by $\sigma^2/n$) to a chi-squared distribution with 1 degree of freedom.

15. Suppose that $X \sim \text{bin}(n_1, p_1)$ and $Y \sim \text{bin}(n_2, p_2)$ and that we wish to test whether the proportions are equal:
$$H_0 : p_1 = p_2 = p \quad \text{versus} \quad H_1 : p_1 \neq p_2$$
where $p$ is unknown.

  a. Give the GLR statistic.
  
  b. Since your GLR does not simplify very well, give an approximate GLRT of size $\alpha$ based on large sample sizes.


**Answer:**

a. The exact GLR statistic is based on the ratio of the likelihoods under $H_0$ and $H_1$.

b. The approximate GLRT for large sample sizes uses the test statistic $Z$ and compares it to the critical value from the standard normal distribution.

15. Let $X_1, X_2, \dots, X_n$ be a random sample from the uniform distribution on $(0, \theta]$. Find the exact (not asymptotic) distribution of $-2 \ln \lambda(\mathbf{X})$ where $\lambda(\mathbf{X})$ is the GLR for 
$$H_0 : \theta = \theta_0 \quad \text{versus} \quad H_1 : \theta \neq \theta_0.$$ 
Based on this, find a GLRT of size $\alpha$.


**Answer:**

Rejects $H_0: \theta = \theta_0$ if:
$$
-2\ln \lambda(X) = 2n\ln\left(\frac{\theta_0}{X_{(n)}}\right) \geq -2\ln(1-\alpha).
$$

This test is exact and does not rely on asymptotic approximations.

9. Consider the distribution with pdf
$$f(x; \theta) = 1 - \theta^2\left(x - \frac{1}{2}\right), \quad 0 < x < 1, \quad -1 < \theta < 1.$$

  a. Find the best test of size $\alpha$ for
  $$H_0 : \theta = 0 \quad \text{versus} \quad H_1 : \theta = \theta_1,$$
  for some fixed $\theta_1 \neq 0$, based on a sample of size 1.
  
  b. Find (if it exists) a UMP (uniformly most powerful) test of size $\alpha$ for
  $$H_0 : \theta = 0 \quad \text{versus} \quad H_1 : \theta \neq 0$$
  based on a sample of size 1.

**Answer:**

a. 
The best test of size $\alpha$ rejects $H_0: \theta = 0$ in favor of $H_1: \theta = \theta_1$ if:
$$
x > 1 - \alpha.
$$

If $\theta_1 < 0$, the test should instead reject $H_0$ if:
$$
x < \alpha.
$$

b. 
The UMP test of size $\alpha$ rejects $H_0: \theta = 0$ if:
$$
x < \frac{\alpha}{2} \quad \text{or} \quad x > 1 - \frac{\alpha}{2}.
$$

This test is uniformly most powerful because it maximizes the probability of rejecting $H_0$ for any nonzero value of $\theta$, while maintaining the size $\alpha$.

14. Consider a random sample $X_1, X_2, \dots, X_n$ from a distribution with pdf 
$$f(x; \theta) = \theta(1 - x)^{\theta-1}, \quad 0 < x < 1, \quad \theta > 0.$$
Give the form of the GLRT (generalized likelihood ratio test) for testing 
$$H_0 : \theta = 1 \quad \text{against} \quad H_1 : \theta \neq 1.$$
You do not need to actually do the test.

**Answer:**

Reject $H_0$ if
$$
2n\log\hat{\theta} + 2(\hat{\theta} - 1)\sum_{i=1}^{n}\log(1 - X_i) > c,
$$

where $\hat{\theta}$ is the MLE of $\theta$ given by
$$
\hat{\theta} = -\frac{n}{\sum_{i=1}^{n}\log(1 - X_i)}.
$$

This statistic is compared against a critical value $c$ that is chosen based on the desired significance level $\alpha$.

8. Let $X$ be a random sample of size 1 from the shifted exponential distribution with rate 1, which has pdf

$$f(x; \theta) = e^{-(x-\theta)} I_{(\theta, \infty)}(x).$$

  a. Find a test of size $\alpha$ for $H_0 : \theta \leq \theta_0$ versus $H_1 : \theta > \theta_0$ based on looking at that single value in the sample.
  
  b. Find the power function for your test.


**Answer:**

a.

Rejects $H_0$ if 
$$
X > \theta_0 - \log(\alpha).
$$

b.

The power function of the test is 
$$
\beta(\theta) = \alpha e^{-(\theta_0 - \theta)}.
$$

18. Let $X_1, X_2, \dots, X_n$ be a random sample from the uniform distribution on the interval $(0, \theta)$. Consider testing

$$H_0 : \theta = \theta_0$$  
$$H_1 : \theta < \theta_0.$$

  a. Find a test of size $\alpha$ based on $X_{(n)}$, the maximum of the sample.
  
  b. Find a test of size $\alpha$ based on $X_{(1)}$, the minimum of the sample.



**Answer:**


a.

The test of size $\alpha$ based on the maximum $X_{(n)}$ is to reject $H_0$ if 
$$
X_{(n)} \leq \theta_0 \cdot \alpha^{1/n}.
$$

b.

The test of size $\alpha$ based on the minimum $X_{(1)}$ is to reject $H_0$ if 
$$
X_{(1)} \geq \theta_0 \cdot (1 - \alpha^{1/n}).
$$
