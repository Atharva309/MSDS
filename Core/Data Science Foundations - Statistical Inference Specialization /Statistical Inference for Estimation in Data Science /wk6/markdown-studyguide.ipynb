## Week 2

1. Let $X_1, X_2, \ldots, X_n$ be a random sample from the Poisson distribution with parameter $\lambda$.

   Which of the following is the maximum likelihood estimator for $\lambda$?

   - [x] The sample mean $\overline{X}$.


2. Suppose that $X_1, X_2, \ldots, X_n$ is a random sample from the $\Gamma(\alpha, \beta)$ distribution where $\alpha$ is known.

   Which of the following is the maximum likelihood estimator for $\beta$?

   - [x] $\frac{\alpha}{\overline{X}}$

3. Consider a random sample $X_1, X_2, \ldots, X_n$ from the shifted exponential distribution with pdf

   $ \
   f(x; \theta) = 
   \begin{cases} 
   e^{-(x - \theta)} & , x \geq \theta \\ 
   0 & , x < \theta 
   \end{cases}
   \ $

   Which of the following is the maximum likelihood estimator for $\theta$?

   - [x] $\min(X_1, X_2, \ldots, X_n)$

4. Consider a random sample $X_1, X_2, \ldots, X_n$ from the continuous Pareto distribution with pdf

   $ \
   f(x; \gamma) = 
   \begin{cases} 
   \frac{\gamma}{(1 + x)^{\gamma + 1}} & , x \geq 0 \\ 
   0 & , x < 0 
   \end{cases}
   \ $

   Here $\gamma > 0$ is a parameter to be estimated.

   Which of the following is the maximum likelihood estimator for $\gamma$?

   - [x] $\frac{n}{\sum_{i=1}^{n} \ln(1 + X_i)}$


5. Suppose that $X_1, X_2, \ldots, X_n$ is a random sample from the shifted exponential distribution with rate $\lambda$ which has pdf

   $$
   f(x; \lambda, \theta) = 
   \begin{cases} 
   \lambda e^{-\lambda (x - \theta)} & , x \geq \theta \\ 
   0 & , x < \theta 
   \end{cases}
   $$

   Which of the following are the MLEs for $\lambda$ and $\theta$?

   - [x] $\hat{\lambda} = \frac{n}{\sum_{i=1}^{n} (X_i - X_{(1)})}$ and $\hat{\theta} = X_{(1)}$ where $X_{(1)} = \min(X_1, X_2, \ldots, X_n)$



## Invariance, Mean-Squared Error, and Efficiency

1. Suppose that $X_1, X_2, \ldots, X_n$ is a random sample from the exponential distribution with rate $\lambda > 0$.

   Find the MLE for $P(X > 1)$ where $X$ is a future sampled value.

   - [x] $e^{-1/\overline{X}}$

2. Let $X_1, X_2, \ldots, X_n$ be a random sample from any distribution with mean $\mu$ and variance $\sigma^2$.

   Consider the following two unbiased estimators of $\mu$.

   $$
   \hat{\mu}_1 = \overline{X}
   $$

   and

   $$
   \hat{\mu}_2 = \frac{X_1}{3} + \frac{2(X_2 + X_3 + \cdots + X_n)}{3(n - 1)}
   $$

   Find the efficiency of $\hat{\mu}_1$ relative to $\hat{\mu}_2$.

   - [x] $\frac{n^2 + 3n}{9(n - 1)}$


3. Suppose that $X_1, X_2, \ldots, X_n$ is a random sample from any distribution with mean $\mu$ and variance $\sigma^2$.

   Consider the estimator of $\sigma^2$ given by

   $$
   \tilde{\sigma}^2 = \frac{\sum_{i=1}^{n} (X_i - \overline{X})^2}{n}.
   $$

   What is the bias of this estimator?

   - [x] $-\frac{1}{n} \sigma^2$

4. The median of a continuous distribution corresponding to a random variable $X$, is a number $\xi$ such that $P(X \leq \xi) = 1/2$.

   If $X_1, X_2, \ldots, X_n$ is a random sample from the distribution with pdf

   $$
   f(x) = 
   \begin{cases} 
   \frac{3x^2}{\theta^3} & , 0 < x < \theta \\ 
   0 & , \text{otherwise}
   \end{cases}
   $$

   find the MLE for the median of the distribution.

   - [x] $\sqrt[3]{1/2} \cdot \max(X_1, X_2, \ldots, X_n)$

5. Let $X_1, X_2, \ldots, X_n$ be a random sample from the Poisson distribution with parameter $\lambda > 0$. Let $\hat{\theta}$ be the MLE for $\lambda$.

   What is the mean-squared error of $\hat{\lambda}$ as an estimator of $\lambda$?

   - [x] $\frac{\lambda}{n}$


## Week 3

1. Let $X_1, X_2, \ldots, X_n$ be a random sample from the normal distribution with mean $\mu$ and variance $\sigma^2$. Assume that $\sigma^2$ is known.

   Find the Cramer-Rao lower bound for the variance of all unbiased estimators of $\mu$.

   - [x] $\frac{\sigma^2}{n}$

2. Let $X_1, X_2, \ldots, X_n$ be a random sample from the normal distribution with mean $\mu$ and variance $\sigma^2$. Assume that $\mu$ is known.

   Find the Cramer-Rao lower bound for the variance of all unbiased estimators of $\sigma^2$.

   - [x] $\frac{2(\sigma^2)^2}{n}$

3. Suppose that $X_1, X_2, \ldots, X_n$ be a random sample from the Poisson distribution with parameter $\lambda > 0$.

   Find the Cramer-Rao lower bound for the variance of all unbiased estimators of $P(X_i = 0)$.

   - [x] $\frac{\lambda e^{-2\lambda}}{n}$

## Further Computations with MLEs
1. Let $X_1, X_2, \ldots, X_n$ be a random sample from the exponential distribution with rate $\lambda > 0$.

   Let $\hat{\lambda}$ be the MLE for $\lambda$.

   Compute the efficiency of $\hat{\lambda}$.

   - [x] $\frac{(n-1)^2(n-2)}{n^3}$


2. Let $X_1, X_2, \ldots, X_n$ be a random sample from the exponential distribution with rate $\lambda$.

   Find the asymptotic distribution of the maximum likelihood estimator for the "tail probability" $P(X > 1)$ for some fixed $x > 0$.

   (Note: You do not actually need to find the MLE itself!)

   - [x] $N(e^{-\lambda}, \lambda^2 e^{-2\lambda} / n)$

3. Let $\hat{\gamma}_n$ be the MLE for the parameter of the Pareto distribution with pdf

   $$
   f(x; \gamma) = 
   \begin{cases} 
   \frac{\gamma}{(1 + x)^{\gamma + 1}} & , x > 0 \\ 
   0 & , \text{otherwise}
   \end{cases}
   $$

   Which of the following are true about $\hat{\gamma}_n$?

   (Check all that apply.)

   - [x] $\lim_{n \to \infty} E[\hat{\gamma}_n] = \gamma$
   - [x] $\hat{\gamma}_n$ has an asymptotically normal distribution

4. Let $X_1, X_2, \ldots, X_n$ be a random sample from any distribution with mean $\mu$ and finite variance $\sigma^2$.

   What is the asymptotic distribution of $\sum_{i=1}^{n} X_i$?

   - [x] $N(n\mu, n\sigma^2)$


5. Let $X_1, X_2, \ldots, X_n$ be a random sample from the distribution with pdf

   $$
   f(x) = \theta x^{\theta - 1} I_{(0,1)}(x).
   $$

   To what does the sample mean $\overline{X}$ converge in probability?

   - [x] $\frac{\theta}{\theta + 1}$

   Give a 95% confidence interval for true average viscosity.

   Calculate an approximate 90% confidence interval for true average fracture strength.

3. Suppose that a random sample of 42 bottles of a particular brand of wine is selected and the alcohol content of each bottle is determined.    Which of the following is true when using the same data set. (Select all that apply.)

   Derive an 80% confidence interval for the true population mean length $\mu$.

   - [x] (21.85, 25.55)


5. You are planning on taking a random sample from a normal distribution with variance 9 and making a 95% confidence interval for its mean $\mu$.

## Confidence Intervals for Differences Between Means

1. A manufacturer of a premium laminate and wood composite flooring system would like to be able to say that their product will experience less warpage than hardwood flooring under certain fixed temperature and humidity conditions.

   A random sample of size 10 from the laminate system produced a sample mean warpage of 2.7 degrees and a random sample of size 12 from the hardwood flooring gave a sample mean warpage of 3.2 degrees. It is believed that both systems have a true warpage variance of 0.8 degrees squared.

   Assuming warpage for both products are normally distributed, find a 90% confidence interval for $\mu_1 - \mu_2$ where $\mu_1$ is the true mean warpage for laminate and $\mu_2$ is the true mean warpage for hardwood.

   - [x] $(-1.130, 0.130)$

2. A manufacturer of a premium laminate and wood composite flooring system would like to be able to say that their product will experience less warpage than hardwood flooring under certain fixed temperature and humidity conditions.

   A random sample of size 10 from the laminate system produced a sample mean warpage of 2.7 degrees and a sample variance of 0.73. A random sample of size 12 from the hardwood flooring gave a sample mean warpage of 3.2 degrees and a sample variance of 0.80. It is believed that the true variance for the laminate system is the same as the true variance for the hardwood system.

   Assuming warpage for both products are normally distributed, find a 90% confidence interval for $\mu_1 - \mu_2$ where $\mu_1$ is the true mean warpage for laminate and $\mu_2$ is the true mean warpage for hardwood.

   - [x] $(-1.147, 0.147)$
  

3. The Slant Shear Test (SST) is a widely used test for evaluating the bond of resinous repair materials to concrete. It was applied to two random samples of concrete slabs which were finished in different ways. Population 1 has a smooth finish while Population 2 has a more textured finish.

   The sample mean shear strength (in N/mm$^2$) and sample variance for the smooth finish sample of size 138 were 18.17 and 1.78, respectively. The sample mean and variance in shear strength for a random sample of size 110 hand-chiseled specimens were 21.66 and 3.21.

   Find a 95% confidence interval $\mu_1 - \mu_2$, the true mean difference in shear strengths.

   - [x] $(-3.892, -3.088)$

## Build Your Own Confidence Intervals

1. Let $X_1, X_2, \ldots, X_n$ be a random sample from the continuous uniform distribution on the interval from 0 to $\theta$ with $\theta > 0$.

   Using the "cdf method" from this lesson, we can show that $Y_n = \max(X_1, X_2, \ldots, X_n)$ has pdf given by

   $$
   f(y; \theta) = 
   \begin{cases} 
   \frac{n}{\theta^n} y^{n-1} & , 0 < y < \theta \\ 
   0 & , \text{otherwise}
   \end{cases}
   $$

   Which of the following is a valid $100(1 - \alpha)\%$ confidence interval for $\theta$?

   (Hint: You might want to look at Lesson 3 of Module 1 to figure out how to form a pivotal quantity.)

   - [x] $\left( \frac{Y_n}{(1 - \alpha)^{1/n}}, \infty \right)$
  

2. Let $X_1, X_2, \ldots, X_n$ be a random sample from the shifted continuous exponential rate 1 distribution with pdf

   $$
   f(x; \theta) = 
   \begin{cases} 
   e^{-(x - \theta)} & , x \geq \theta \\ 
   0 & , \text{otherwise}
   \end{cases}
   $$

   Which of the following represents a 95% confidence interval for $\theta$ based on the minimum value $Y_n = \min(X_1, X_2, \ldots, X_n)$.

   - [x] $\left( Y_n + \frac{\ln 0.05}{n}, Y_n \right)$


3. Let $X_1$ be a random sample of size 1 from the $\Gamma(3, \beta)$ distribution with parameter $\beta > 0$.

   Derive a 90% confidence interval for $\beta$ based on the single observation $X_1$.

   (Yes, a sample of size 1 is silly but it makes for an easier computation and the point of this problem is to test the techniques for developing confidence intervals!)

   - [x] $\left( 0, \frac{\chi^2_{0.10, 6}}{2X_1} \right)$
