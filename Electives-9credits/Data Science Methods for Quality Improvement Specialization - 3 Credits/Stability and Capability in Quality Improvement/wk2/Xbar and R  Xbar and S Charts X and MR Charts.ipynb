```python
import pandas as pd

# Read the data from BagWeight.dat
df = pd.read_csv('Bag_Weight.dat', delim_whitespace=True)

# Calculate X-bar and R for each week
xbar = df.groupby('Week')['Bag_Weight'].mean()
ranges = df.groupby('Week')['Bag_Weight'].apply(lambda x: x.max() - x.min())

# Calculate X-double bar and R-bar
x_double_bar = xbar.mean()
r_bar = ranges.mean()

# Constants for n=5
A2 = 0.577
D3 = 0
D4 = 2.114

# Calculate UCL and LCL for X-bar chart
UCL_xbar = x_double_bar + A2 * r_bar
LCL_xbar = x_double_bar - A2 * r_bar

# Calculate UCL for R chart
UCL_r = D4 * r_bar

# Display the results
print(f"Upper Control Limit for X-bar Chart: {UCL_xbar:.4f}")
print(f"Lower Control Limit for X-bar Chart: {LCL_xbar:.4f}")
print(f"Upper Control Limit for R Chart: {UCL_r:.4f}")
```

```python
import pandas as pd

# Read the data from ShearedSheet.dat
df = pd.read_csv('ShearedSheet.dat', delim_whitespace=True)

# Calculate X-bar and R for each subgroup
xbar = df.groupby('Subgroup')['Width_Deviation'].mean()
ranges = df.groupby('Subgroup')['Width_Deviation'].apply(lambda x: x.max() - x.min())

# Calculate X-double bar and R-bar
x_double_bar = xbar.mean()
r_bar = ranges.mean()

# Constants for n=8
A2 = 0.373
D3 = 0.136
D4 = 1.864

# Calculate UCL and LCL for X-bar chart
UCL_xbar = x_double_bar + A2 * r_bar
LCL_xbar = x_double_bar - A2 * r_bar

# Calculate UCL and LCL for R chart
UCL_r = D4 * r_bar
LCL_r = D3 * r_bar

# Display the results
print(f"Upper Control Limit for X-bar Chart: {UCL_xbar:.4f}")
print(f"Lower Control Limit for X-bar Chart: {LCL_xbar:.4f}")
print(f"Upper Control Limit for R Chart: {UCL_r:.4f}")

# Handle LCL for R chart separately
if LCL_r > 0:
    print(f"Lower Control Limit for R Chart: {LCL_r:.4f}")
else:
    print("Lower Control Limit for R Chart: NA")
```
