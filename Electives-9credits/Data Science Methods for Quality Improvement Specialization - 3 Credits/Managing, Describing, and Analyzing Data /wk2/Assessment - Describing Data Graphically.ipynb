```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("viscosity.txt")
data     
```

```python

# Generate the grouped frequency distribution
bins = [0, 20, 40, 60, 80, 100]  # Define the bin intervals
frequency = pd.cut(data['centistokes'], bins=bins, include_lowest=True).value_counts().sort_index()
midpoints = [(bins[i] + bins[i+1]) / 2 for i in range(len(bins)-1)]  # Calculate the midpoints

# Print the grouped frequency distribution
print('Grouped Frequency Distribution:')
for i in range(len(frequency)):
    print(f'{frequency.index[i]}: {frequency[i]}')

# Generate the grouped histogram
plt.figure(figsize=(8, 6))
plt.hist(data['centistokes'], bins=bins, edgecolor='black')
plt.xlabel('centistokes')
plt.ylabel('Frequency')
plt.title('Grouped Histogram of Viscosity')
plt.xticks(midpoints)  # Set the x-axis tick locations to the midpoints
plt.grid(True)
plt.show()

# Calculate the five-number summary
summary = data.describe().loc[['min', '25%', '50%', '75%', 'max']]
print('\nFive Number Summary:')
print(summary)

# Generate the box and whisker plot
plt.figure(figsize=(8, 6))
plt.boxplot(data['centistokes'], vert=False)
plt.xlabel('centistokes')
plt.title('Box and Whisker Plot of Viscosity')
plt.grid(True)
plt.show()
```

```python
import numpy as np

hist, edges = np.histogram(data, bins=10)  # You can adjust the number of bins as needed
relative_frequencies = hist / np.sum(hist)

highest_relative_frequency = max(relative_frequencies)
print(f"Highest Relative Frequency: {highest_relative_frequency:.2f}")
```
