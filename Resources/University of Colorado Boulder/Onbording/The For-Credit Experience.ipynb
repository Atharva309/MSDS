## The For-Credit Experience

Our for-credit courses offer a different experience than our non-credit courses, including:

- Set dates:  For-credit courses have session start and end dates. Each session is 8 weeks long.   See the calendar for exact dates. All for-credit coursework must be submitted by the last day of the 8-week session, and tuition cannot be carried over to future terms.

- CU tuition: For-credit tuition is paid to the University of Colorado Boulder and Includes access to additional CU services. (e.g., online library and alumni/career services, CU on Coursera membership, and an @colorado.edu email address).

- Academic support: Course facilitators support for-credit courses through weekly office hours, discussion forums, and email.

- CU credit: Students receive academic credit for their for-credit courses on a transcript from the University of Colorado Boulder.

You must unlock the final exam to earn a course grade. 

To earn a grade for your course, you must pass the password quiz (or equivalent, as an honor code verification) that unlocks your final exam/project before the end of the session. 

- This is a required step to earn a course grade even if you have completed all other coursework and/or decide to skip the final exam/project for the course. 

- If you do not pass the password quiz (or equivalent), you will be withdrawn after the session ends and earn an administrative W for the course regardless of your current course grade. 

Furthermore:

You will complete the password quiz to access restricted content. Once you click “Mark as completed” below, you will not be eligible to drop or withdraw from a class or receive a refund even if you do not access the final exam/project itself. 

DO NOT open the following quiz item, "Honor Code Verification," until you are ready to submit your final project. Responding to and submitting the verification will unlock the project. Once you do this, you will no longer be eligible for a tuition refund.

## Program Overview, Academic Experience, and Student Handbook

#### Welcome 

Welcome to the University of Colorado Boulder’s Master of Science degree in Data Science (MS-DS), a cutting-edge program designed for the twenty-first century. Hosted on the internationally acclaimed Coursera

 learning platform, the online MS-DS program is a fully accredited Master’s Degree in Data Science. There are no qualifying designations on official CU Boulder transcripts, degrees, or certificates associated with it. 

Please feel free to reach out to msds-support@colorado.edu

 with questions. 

We are excited for you to join our community and forge a new kind of education for the twenty-first century!

#### Vision

The MS-DS is founded on the principles of the Massive Open Online Course (MOOC) “revolution.” In 2012 a number of major research universities developed online teaching platforms that allowed tens of thousands of students to enroll in a single course. Because MOOC technology allowed greater access to education, it was widely heralded as a revolution in higher education that would disrupt the ways degrees are offered to the world.

Recognizing this, the Departments of Applied Mathematics, Computer Science, Information Science, and others at the University of Colorado Boulder developed the online MS-DS degree program with a new vision for graduate education.

The degree presents three major innovations:

1. Access 

The MS-DS is designed to provide global access to graduate-level education. It seeks to educate the world and uses performance-based admissions rather than traditional admissions standards. Anyone who is able to do graduate-level data science coursework is welcome in our degree. There are two key steps in the performance-based admissions process in order to be admitted:

- Declare your intention to pursue the degree, which can happen at any time during your academic journey through the MS-DS curriculum.

- Complete the required courses in one of several designated “pathway specializations” with a grade of B or better and maintain an overall grade point average of 3.0 (B) or better. This GPA level must be maintained across the designated “pathway” courses and is cumulative for all for-credit online MS-DS courses taken previously. 

2. Curriculum 

The MS-DS curriculum is both modular and self-directed. The faculty have purposely broken down semester-long courses into small modules that naturally fit together in groups. 

The details of the MS-DS curriculum are outlined on the MS-DS on Coursera Curriculum page and highlighted below.

3. Learning 

The online MS-DS is guided by the belief that learning belongs to the learner. To be successful in the MS-DS, you must commit to your learning by creating a clear plan of courses, a strategy for taking those courses, and a schedule for study.

You may explore the non-credit versions of the MS-DS course before taking them for credit to see if the course content fits your learning plan.
How does it all work?

There is additional detailed program information in the online MS-DS student handbook
 and on the MS-DS website. Additionally, below is a high-level overview of some program features.

#### Tuition

- Cost per credit hour: $667

- Tuition for the full master's degree: $20,010

- Pay as you go – charged per credit hour at time of enrollment

- This program is not currently eligible for U.S. federal financial aid

- There are currently no scholarships available

- Some course materials can be previewed at no cost

#### Courses and Specializations

- Courses are open for 8 weeks – see the calendar

- Some courses in the curriculum are grouped into specializations of three courses

- Specializations are a series of courses that cover specific topics within a focus area (Statistical Inference for Data Science, Data Structures & Algorithms, etc.)

- A pathway specialization is one that is designated for entry into the degree program

- Degree-seeking students must successfully complete all of the for-credit courses within one "pathway" specialization with a B or better to be admitted to the program

- You can earn graduate academic credit for each individual course in the curriculum

#### Sessions

- There are 6 eight-week sessions per year – see the calendar

- The session enrollment period starts 2 weeks before the first day course content is available

- The session enrollment period ends 2 weeks before all coursework is due

- All for-credit coursework is due before the last day of each session

- While work can be done at your own pace within the session, the for-credit session end date is absolute

#### Preview Courses with Non-Credit Options

- Each course in the MS-DS can be previewed or taken as a non-credit Coursera class

- You do not pay CU tuition when previewing a class but may need to pay fees to Coursera for access to a full range of content

- You may upgrade from non-credit to for-credit at any time during your learning journey. Previously completed assignments will be automatically applied to your for-credit experience

- Additional material and assessments must be completed to earn credit; these materials are only available when you pay tuition

#### Discussion Forums

Our program staff are here to support your learning. If you have any questions about the content covered in a course, please post it in the discussion forum. We will get back to you within 24–48 business hours. By posting in the forum, you are also letting other students benefit from the response. More information about where to find and how to use discussion forums will be covered in the Coursera platform information lesson in this onboarding course.

Reach out to msds-support@colorado.edu if you have questions regarding:

- Your grades or assignments

- Degree planning

- Course registration

- Tuition or fees

- Request for official and unofficial transcripts

- A private issue that only affects you
