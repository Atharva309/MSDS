```r
loaded_dice <- function(max_rolls = 50000, sides = 8, num_of_dice = 3) {
    
  # probabilities of each side
  p <- c(rep(0.1, sides - 1), 0.3) 
  rolls <- replicate(num_of_dice, sample(sides, max_rolls, replace = TRUE, prob = p), simplify = FALSE)
  sums <- rowSums(matrix(unlist(rolls), ncol = num_of_dice, byrow = TRUE))
  
  hist(sums, breaks = seq(2, num_of_dice * sides, by = 1), col = "lightblue",
       xlab = "Sum of dice rolls", main = "Histogram of 50,000 rolls of three 8-sided dice")
}
```

```r
loaded_dice(max_rolls = 100000)
```

```r
 both_na <- function(x, y) {
  sum(is.na(x) & is.na(y))
}
```
