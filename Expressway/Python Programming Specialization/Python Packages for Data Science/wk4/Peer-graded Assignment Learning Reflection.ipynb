```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import sklearn
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
housing = pd.read_csv("../housing/housing.csv")
```

```python
housing
```

```python

pairplot = housing[[ "median_income", "total_rooms",
"housing_median_age"]]
sns.pairplot(pairplot, hue='median_income')
```

```python
sns.relplot(
    data=housing, kind="line",
    x=housing["median_income"], y=housing["housing_median_age"],
)
```

```python
housing.plot(kind="scatter", x="longitude", y="latitude", alpha=0.4,
s=housing["population"]/100, label="population", figsize=(10,7),
c="median_house_value", cmap=plt.get_cmap("jet"), colorbar=True,
)
plt.legend()
```
