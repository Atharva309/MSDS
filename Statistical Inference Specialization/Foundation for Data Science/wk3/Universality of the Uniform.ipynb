```r
#set grid
par(mfrow = c(1,2))

#Exponential
hist(rexp(1000), col = rgb(1, 0, 0, 1/2),
     main = "Exponential Distribution",
     xlab = "")

#transformed Uniform
hist(-log(1 - runif(1000)), col = rgb(0, 1, 0, 1/2),
     main = "-ln(1 - U)", xlab = "")
```

```r
#generate the r.v., an Expo(1)
X = rexp(1000, 1)

#plot F(X)
hist(pexp(X, 1), col = rgb(0, 0, 1, 1/2),
     main = "F(X)", xlab = "")
```
